# Loading a kappa map

First, load modules for loading a map of hdf5 format

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

Set parameters relevant to the map

```python
ng = 1024  # number of grids
theta = 5.0  # opening angle in deg
fname = "sample_map.hdf5"  # This sample file contains the kappa map with zs = 1.034

pix = theta/ng  # pixel size
theta = pix*np.arange(ng)
```

Load the sample map file with `h5py`

```python
with h5py.File(fname, "r") as f:
    kappa = f["z23/kappa"][:]

print(kappa, kappa.shape)
```

The array `kappa` contains the convergence field at $z_s = 1.034$. Let's plot it as the 2D color map.

```python
X, Y = np.meshgrid(theta, theta)

plt.pcolormesh(X, Y, kappa, vmin=-0.05, vmax=0.07, cmap="Blues")
plt.gca().set_aspect('equal')
cbar = plt.colorbar()
cbar.set_label(r'$\kappa$', fontsize=20)
```
